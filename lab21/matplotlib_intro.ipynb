## Basic Concepts

Reference: [Parts of a Figure](https://matplotlib.org/stable/tutorials/introductory/quick_start.html#parts-of-a-figure)

### Figure

The top level container for all the plot elements. A Figure Object contains the whole plot.
You can have more than one Figure object.
Reference: [Figure Class](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure)


### Axes (and Subplots)
The Axes represent an **individual drawing plot**, with its own coordinate system.

**A figure can contain many Axes, but a given Axes object can only be placed in one Figure.**

#### Create figure and axes with Pyplot Interface: 

```python
fig = plt.figure()
fig.set_facecolor("#B15A5A")

# add a new axes at the position 1 on a grid of 2 rows and 1 column.
fig.add_subplot(211)

# add a new axes at the position 2 on a grid of 2 rows and 1 column.
fig.add_subplot(212)

# Show the plot
plt.show()
```

#### Create figure and axes using OOP Interface

This is the preferred way of creatting fugure and axes objects.

```python
fig, axes = plt.subplots(nrows=2)
fig.set_facecolor("#B15A5A")
```

### Axes title and labels

Each Axes can have a title ([Axes.set_title](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html#matplotlib.axes.Axes.set_title)) and labels ([Axes.set_xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html), [Axes.set_ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html#matplotlib.axes.Axes.set_ylabel)) 

```python
fig, axes = plt.subplots()
fig.set_facecolor("#B15A5A")

axes.set_title("y=sqrt(x)")

axes.set_xlabel("X",fontsize=14, color="yellow")
axes.set_ylabel("Y", rotation=0,fontsize=14, color="yellow")

# move labels position
axes.xaxis.set_label_coords(1.05, 0.02)
axes.yaxis.set_label_coords(0, 1.05)

# Show the plot
plt.show()
```

### Ticks

Each axis has a ticks. With Matplotlib you can configure all aspects (location, formating, etc.) of the ticks

```python
fig, axes = plt.subplots(nrows=1)
fig.set_facecolor("#B15A5A")

axes.get_xticklabels()[3].set_color("#F3FF00")

# Show the plot
plt.show()
```

### Legend

In Matplotlib, a legend is an area describing the elements of the graph. We can place a legend on an exes using the [matplotlib.pyplot.legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) method.

The legend will be placed in the best location (upper left) by default. You could specify location of the legend using loc argument in legend() function.

	`ax.legend(loc='upper left')`

You could also change the style of your legend using various options like font size, label color, background color etc.

	`ax.legend(loc = 'upper left', fontsize = 'large', shadow=True, framealpha = 0.5)`


Here is an example of how you can create a simple line plot with a legend in Matplotlib:

```python
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y1, label='Line 1')
ax.plot(x, y2, label='Line 2')

# Add a legend
# ax.legend()
# ax.legend(loc='upper left')
ax.legend(loc = 'upper left', fontsize = 'large', shadow=True, framealpha = 0.5)

# Show the plot
plt.show()
```

## Matplotlib APIs (Pyplot vs OOP Interface)

Matplotlib offers two main ways to create visualizations: 

1. pyplot interface (procedural interface).
1. Object-Oriented interface (OO).

It is very important to remember that **you should not mix both approaches**!

Reference: [pyplot-vs-object-oriented-interface](https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/)

### State-based (implicit) API:

This is the older approach! 

The state-based API uses a global state (pyplot) to manage the current figure and axes, and all plotting commands are applied to this current object.

State-based API is mainly intended for interactive plots and simple cases of programmatic plot generation, and is convenient for users familiar with MATLAB.

The basic workflow of creating a visualization with the state-based API is to first create a figure and axes, and then use various plotting functions (such as plot, scatter, and hist) to add data and other elements to the figure. 

**Each function call modifies the current state of the plot, so the order of the function calls is important!**

For example:

```python
#  Plot line using State-based interface (pyplot api):
x = np.linspace(0,10,100)

# plot on the 1st axes:
plt.subplot(2, 1, 1)
plt.title('sin')
plt.plot(x, np.sin(x), color="#F00")

# plot on the 2nd axes:
plt.subplot(2, 1, 2)
plt.title('cos')
plt.plot(x, np.cos(x), color="#0F0")
```

### OOP (explicit) API

The OOP interface is more powerful and flexible. 

It allows you to create plots by creating instances of different classes, such as Figure, Axes, and Line2D, that represent the different parts of the plot. 

**The class instances can be manipulated and modified directly, so the order of the function calls is less important.**

For example:

```python
#  Plot line using OOP interface:
x = np.linspace(0,10,100)

# initialize figure and axes objects:
fig, [ax1,ax2] = plt.subplots(2)

# plot on the 1st axes:
ax1.set_title('sin')
ax1.plot(x, np.sin(x), color="#F00")

# plot on the 2nd axes:
ax2.set_title('cos')
ax2.plot(x, np.cos(x), color="#0F0")
```

## Multiple plots in single axes

We can add 2 or more plots in single axes, as well. Check next examples.

### With State-based API:

```python
n=100
x = np.linspace(-np.pi,np.pi,n)
y = np.sin(2*x)

plt.plot (x, y+1, color='blue', alpha=0.5)
plt.plot (x, y-1, color=(0,1,0,0.2)) # equiv to: color='green', alpha=0.2
```

### With OOP API:

```python
n = 100
x = np.linspace(-np.pi,np.pi,n)
y = np.sin(2*x)

fig, ax = plt.subplots()

ax.plot (x, y+1, color='blue', alpha=0.5)
ax.plot (x, y-1, color=(0,1,0,0.2)) # equiv to: color='green', alpha=0.2
```

## Controlling style of text and labels using a dictionary

Matplotlib provides many properties to control every aspect of text and its layout: [Text properties and layout](https://matplotlib.org/stable/tutorials/text/text_props.html)
<!-- 
- [fontdict example](https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_fontdict.html#sphx-glr-gallery-text-labels-and-annotations-text-fontdict-py) -->

```python
fig, axes = plt.subplots()
fig.set_facecolor("#B15A5A")

axes.set_title("y=sqrt(x)",
	fontdict={
		'fontname': 'monospace', # try 'cursive' or 'serif'
		'fontsize': 30,
		'fontweight': 'bold',
		'color': 'red',
		'backgroundcolor':'#999',
		'verticalalignment': 'bottom',
	 },
	 loc='right'
)

axes.set_xlabel("X",fontsize=14, color="yellow")
axes.set_ylabel("Y", rotation=0,fontsize=14, color="yellow")

axes.xaxis.set_label_coords(1.05, 0.02)
axes.yaxis.set_label_coords(0, 1.05)

# Show the plot
plt.show()
```

## Set color values

Reference: [Specifying Colors](https://matplotlib.org/stable/tutorials/colors/colors.html#sphx-glr-tutorials-colors-colors-py)

In matplotlib, you can specify colors using a variety of different ways:

1. Single letter codes: You can specify colors using single letter codes, such as 'b' for blue, 'g' for green, 'r' for red, 'c' for cyan, 'm' for magenta, 'y' for yellow, 'k' for black, and 'w' for white. For example, to plot a line in yellow, you can use the following code:

	`plt.plot(x, y, color='y')`

2. RGB tuple: You can specify colors using an RGB (Red, Green, Blue) tuple, where each element of the tuple is an integer between 0 and 1. For example, the color red can be specified using the tuple (1, 0, 0), green can be specified using the tuple (0, 1, 0), and blue can be specified using the tuple (0, 0, 1). And by mixing these base colors you can prodice any color. Yellow, for instance is a combination of Red and Green, so the tuple will be: (1, 1, 0)

	For example, to plot a line in yellow, you can use the following code:
	
	`plt.plot(x, y, color=(1, 1, 0))`

3. Hex code: You can also specify colors using a hex code. It's a RGB value represented in hexadecimal format. The code #FF0000 (or #F00) correspond to red color, #00FF00 (or #0F0) - to green, and #0000FF (or #00F) - to blue.
   
   For example, to plot a line in yellow, you can use the following code:
	
	`plt.plot(x, y, color='#FF0')`

4. HTML names: This method is similar to Single letter codes, it allows you to specify any color using its common name, it's very intuitive and simple to use.

	For example, to plot a line in yellow, you can use the following code:
	
	`plt.plot(x, y, color='yellow')`
   

```python
x = np.arange(5)
y = x**2

fig, axes=plt.subplots()
axes.plot(x, y, color='yellow')

# Show the plot
plt.show()
```

## Formatting the plot style

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot.

The format string is a shorthand way of specifying the color, linewidth, and linestyle of a line in a single string. The general syntax for a format string is:

	'[marker][line][color]', 
	
where each of the three elements is optional. 

The default format string is 'b-', which is a solid blue line.

For deeper explanation of format string check [Notes on matplotlib.pyplot.plot docs](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#:~:text=Notes,-Format%20Strings)

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

fig, axes=plt.subplots()
axes.plot(x, y, '*--g')
# '*' stands for star marker,
# '--' stands for dashed line style
# 'g' stands for green color

plt.show()
```

### Changing the linewidth and marker size

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

fig, axes=plt.subplots()
axes.plot(x, y, '*--g', linewidth=5, markersize=20)
# '*' stands for star marker,
# '--' stands for dashed line style
# 'g' stands for green color
plt.show()
```

## Lets put it all together

```python
x = np.array([1,2,3,4])
y1 = x**2
y2 = x**3

fig, axes = plt.subplots(nrows=2)
fig.set_facecolor("#CCCCCC")

axes_names = ['Axes 1', 'Axes 2']
axes_label = ['x**2', 'x**3']
axes_colors = ["#0F0","#00F"]

for ax, y, name, label, color in zip(axes, [y1, y2], axes_names, axes_label, axes_colors ):
    ax.plot(x, y, color=color, label=label)
    ax.set_title(name)
    ax.set_facecolor("#BA5E5E")
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis', rotation="horizontal", labelpad=20.0)
    ax.legend()

axes[0].get_xticklabels()[2].set_color("#F3FF00")
axes[1].get_xticklabels()[2].set_color("#00F3FF")

plt.tight_layout()
```
