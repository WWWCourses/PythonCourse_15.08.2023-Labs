```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some data to be used in next examples:
x = np.linspace(1,100,10)
x
```

## add_axes()

```python
fig = plt.figure()
ax1 = fig.add_axes((0, 0, 1, 1))
ax2 = fig.add_axes((0.5, 0.5, 0.5, 0.5))
ax3 = fig.add_axes((0, 0.25, 0.5, 0.5))
```

```python
ax1.plot(x, x**2, 'r--')
ax2.plot(x, 5+x**2, color='green', linestyle='dashed' )
ax3.plot(x, (2*x)**2, color='blue', linestyle='dashed' )

fig
```

```python
ax1.set_title('x**2')
ax2.set_title('5+x**2')
ax3.set_title('(2*x)**2')

plt.tight_layout()
fig
```

## add_subplot()

```python
# Create 2 axes, stacked vertically:
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2, sharey=ax1)
ax3 = fig.add_subplot(2,2,(3,4))

ax2.set_ylim([0, 10_000])

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")
ax3.plot(x,2*x**2, color="blue")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')
ax3.set_title('2*x**2')

# adjust layout
plt.tight_layout()
```

```python
# Create 2 axes, stacked vertically:
plt.figure()

plt.subplot2grid([2,2],[0,0])
plt.plot(x,x**2, color="red")
plt.title('x**2')

plt.subplot2grid([2,2],[0,1])
plt.plot(x,x**3, color="green")
plt.ylim([0, 10_000])
plt.title('x**3')

plt.subplot2grid([2,2],[1,0], colspan=2)
plt.plot(x,2*x**2, color="blue")
plt.title('2*x**2')

# # adjust layout
plt.tight_layout()
```

## plt.subplots()

```python
fig, ax = plt.subplots()
print(fig)
print(ax)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1)
print(fig)
print(ax)
```

```python
ax1 = ax[0]
ax2 = ax[1]

ax1.plot(x, x**2)
ax2.plot(x, x**3)
fig
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1, height_ratios=[2/3,1/3])
print(fig)
print(ax1)
print(ax2)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

## fig size

```python
fig,ax = plt.subplots(figsize=(8,4), dpi=100)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig,ax = plt.subplots(figsize=(4,2), dpi=200)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig,ax = plt.subplots(figsize=(16,8), dpi=50)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```
