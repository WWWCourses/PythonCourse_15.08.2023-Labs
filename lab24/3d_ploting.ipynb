## plot_surface()

```python
fig, ax = plt.subplots( subplot_kw={'projection':'3d'} )

# Define data
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
# z = np.array([7, 8, 9])

xx,yy = np.meshgrid(x,y)
zz = xx+yy

# Plot the line
ax.plot_surface(xx, yy, zz, cmap='plasma')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# set ticks
ax.set_xticks(x)
ax.set_yticks(y)
ax.set_zticks(x+y)

ax.set_title('Basic 3D Line Plot')
```

## contour

```python
fig, ax = plt.subplots( subplot_kw={'projection':'3d'} )

# Define data
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
# z = np.array([7, 8, 9])

xx,yy = np.meshgrid(x,y)
zz = xx+yy

# Plot the line
ax.contour(xx, yy, zz, cmap='plasma')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# set ticks
ax.set_xticks(x)
ax.set_yticks(y)
ax.set_zticks(x+y)

ax.set_title('Basic 3D Line Plot')
```
